```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina'

plt.style.use('dark_background')
```

## Exercise: Visualizing a Dataset <a name="first-bullet"></a>

We'll use the Python module [`scikit-learn`](https://scikit-learn.org/stable/) to load the `iris` dataset. This dataset contains the sepal length, sepal width, petal length, and petal width for 150 flowers; each flower is classified as one of three iris types, labeled 0, 1, and 2 in `iris_classification`. You can read more about this dataset [here](https://archive.ics.uci.edu/ml/datasets/iris).

```python
iris = datasets.load_iris()
iris_data = iris['data']
iris_classification = iris['target']

#The features of each flower in the iris dataset:
print(iris['feature_names'])

#As an example, the four characteristics of the first flower can be accessed as follows:
print('\n')
print(iris_data[0])

#As another example, an array of all sepal lengths can be accessed as follows:
print('\n')
print(iris_data[:,0])

#The classification (i.e., the type of iris) of each flower is given by 0, 1, or 2 and is stored in an array:
print('\n')
print(iris_classification)
```

Here are examples of plots you can make using this data. These plots only use "type 0" irises; in the exercises below, you'll explore how features depend on iris type.

```python
iris_data[iris_classification==0][:,3]
```

```python
plt.figure(figsize=(8,6))

plt.scatter(iris_data[iris_classification==0][:,0],iris_data[iris_classification==0][:,1],
            s=100*iris_data[iris_classification==0][:,3],label='Type 0')

plt.scatter(iris_data[iris_classification==1][:,0],iris_data[iris_classification==1][:,1],
            s=100*iris_data[iris_classification==1][:,3],label='Type 1')

plt.scatter(iris_data[iris_classification==2][:,0],iris_data[iris_classification==2][:,1],
            s=100*iris_data[iris_classification==2][:,3],label='Type 2')

plt.xlabel('sepal length (cm)', fontsize=14)
plt.ylabel('sepal width (cm)', fontsize=14)
plt.legend(loc=2, fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.hist(iris_data[iris_classification==0][:,0]/iris_data[iris_classification==0][:,2],alpha=1.0,
            label='Type 0')

plt.hist(iris_data[iris_classification==1][:,0]/iris_data[iris_classification==1][:,2],alpha=1.0,
            label='Type 1')

plt.hist(iris_data[iris_classification==2][:,0]/iris_data[iris_classification==2][:,2],alpha=1.0,
            label='Type 2')

plt.xlabel('sepal length/petal length', fontsize=14)
plt.ylabel('Number', fontsize=14)
plt.legend(loc=2, fontsize=14)

plt.show()
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

ax.scatter(iris_data[iris_classification==0][:,0],
           iris_data[iris_classification==0][:,1],
           iris_data[iris_classification==0][:,2],
          s=100*iris_data[iris_classification==0][:,3],label='Type 0')

ax.scatter(iris_data[iris_classification==1][:,0],
           iris_data[iris_classification==1][:,1],
           iris_data[iris_classification==1][:,2],
          s=100*iris_data[iris_classification==1][:,3],label='Type 1')

ax.scatter(iris_data[iris_classification==2][:,0],
           iris_data[iris_classification==2][:,1],
           iris_data[iris_classification==2][:,2],
          s=100*iris_data[iris_classification==2][:,3],label='Type 2')

ax.set_xlabel('sepal length (cm)',fontsize=14)
ax.set_ylabel('sepal width (cm)',fontsize=14)
ax.set_zlabel('petal length (cm)',fontsize=14)

ax.legend(loc=2,fontsize=16)

plt.show()
```
